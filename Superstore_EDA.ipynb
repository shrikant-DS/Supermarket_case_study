```python
df.rename(columns={'Total': 'Sales'}, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
rounding_cols = ['gross margin percentage', 'gross income', 'Sales', 'Tax 5%', 'cogs']
for i in rounding_cols:
    df[i] = df[i].apply(lambda x: round(x, 2))
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['day'] = df['Date'].dt.day
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year
df['day_of_week'] = df['Date'].dt.day_of_week
```

```python
df.groupby(['Branch'])['Sales'].sum().plot.bar(color= sns.color_palette('Set1'))
```
