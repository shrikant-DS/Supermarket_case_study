```python
rounding_cols = ['gross margin percentage', 'gross income', 'Sales', 'Tax 5%', 'cogs']
for i in rounding_cols:
    df[i] = df[i].apply(lambda x: round(x, 2))
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['day'] = df['Date'].dt.day
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year
df['day_of_week'] = df['Date'].dt.day_of_week
```

```python
# Calculate total sales by product line and visualize
sales_by_product = df.groupby('Product line')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=sales_by_product.values, y=sales_by_product.index, palette='viridis')
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()

# Insight:
print("Insight: The chart above shows which product lines generate the highest total sales. This helps stakeholders identify the most profitable product categories and focus business strategies accordingly.")
```
